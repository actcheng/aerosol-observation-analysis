## Plotting: Specific site

To plot the time series of any specific site, the `single_site_plot`(For AOD) and `single_site_size_plot` (For size information) is used.

### Optical properties


```python
site_plot = ana.plot_single_site(aeronet)
```

**Plot time series**

Different variables and sites can be plotted at the same time. To suppress outputs and save the files instead, set `save=True,show=False`.

```python
sites=['Ilorin','Niamey']
var = ['aod','alfa']
site_plot.plot_sites(sites=sites,var=var,plot_time=True)
```

### Compare two variables

```python
site_plot.plot_sites(sites=sites,var=['aod','alfa'],plot_2var=True,plot_time=False)
```

### Size distributions

```python
site_plot = ana.plot_single_site_size(aeronet)
```

**Average distribution**

```python
site_plot.plot_dvdlnr('Niamey')
```

**Time series**

The number and volume concentrations can be plotted against time. The default plots all available data, but `time_range` can be used to specify the exact time period.

```python
site_plot.plot_dV_time('Niamey')
site_plot.plot_dN_time('Niamey')
```

```python
time_range = [datetime.date(2006,11,1),datetime.date(2006,12,31)]
site_plot.plot_dN_time('Niamey',time_range=time_range)
```
